```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


img_red = cv2.imread("red_dot.jpg", cv2.IMREAD_UNCHANGED)
img_red = cv2.cvtColor(img_red, cv2.COLOR_BGR2RGB)

print(img_red.shape)

with open("red_dot.data", "wb") as f:
    f.write(img_red.ravel())

with open("red_dot.data", "rb") as f:
    teste_leitura = f.read()
teste_leitura = np.frombuffer(teste_leitura, dtype = np.uint8)
```

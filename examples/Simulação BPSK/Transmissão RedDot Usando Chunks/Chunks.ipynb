```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


img_penguin = cv2.imread("penguin.jpg", cv2.IMREAD_UNCHANGED)
img_penguin = cv2.cvtColor(img_penguin, cv2.COLOR_BGR2RGB)

print(img_penguin.shape)

with open("penguin_data.data", "wb") as f:
    f.write(img_penguin.ravel())

with open("penguin_data.data", "rb") as f:
    teste_leitura = f.read()
teste_leitura = np.frombuffer(teste_leitura, dtype = np.uint8)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image
```

```python
with open("red_dot.jpg", "rb") as f:
    _ = f.read()

jpg_orig = np.frombuffer(_, dtype = np.uint8)
jpg_orig.setflags(write = True)
```

```python
# simulando o erro da transmiss√£o do arquivo em formato JPG
jpg_orig[1000] = 14
jpg_orig[500] = 2
jpg_orig[2000] = 200
```

```python
write_path = "red_dot_write.jpg"
```

```python
with open("red_dot_write.jpg", "wb") as f:
    f.write(jpg_orig)
```

```python
write_image = cv2.cvtColor(cv2.imread(write_path), cv2.COLOR_BGR2RGB)

plt.imshow(write_image)
```

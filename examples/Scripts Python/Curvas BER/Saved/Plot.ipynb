```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
theory = np.load("theory_bpsk_63.npy")
```

```python
bpsk_puro = np.load("simu_bpsk_63.npy")
```

```python
theory_coded_63 = np.load("theory_bpsk_coded_63.npy")
theory_coded_15 = np.load("theory_bpsk_coded_15.npy")
```

```python
bpsk_coded_63 = np.load("simu_coded_bpsk_63.npy")
bpsk_coded_15 = np.load("simu_coded_bpsk_15.npy")
```

```python
bpsk_reed_m3_t2 = np.load("bpsk_reed_solomon_m3_t2.npy")
bpsk_reed_m8_t11 = np.load("bpsk_reed_solomon_m8_t11.npy")
```

```python
x_data = np.arange(0, 11, 1)
```

```python
plt.semilogy(x_data, theory, "b-", label = "BPSK Teórico")
plt.semilogy(x_data, bpsk_puro, "b*", label = "BPSK Simulado")

plt.semilogy(x_data, theory_coded_63, "r-.", label = "BPSK Codificado Teórico - BCH(63, 30, 6)")
plt.semilogy(x_data, bpsk_coded_63, "r*", label = "BPSK Codificado Simulado - BCH(63, 30, 6)")

plt.semilogy(x_data, theory_coded_15, "g-.", label = "BPSK Codificado Teórico - BCH(15, 7, 2)")
plt.semilogy(x_data, bpsk_coded_15, "g*", label = "BPSK Codificado Simulado - BCH(15, 7, 2)")

plt.semilogy(x_data, bpsk_reed_m3_t2, "*-", color = "purple", label = "BPSK Codificado Simulado - ReedSolomon ($m = 3$, $t = 2$)")
plt.semilogy(x_data, bpsk_reed_m8_t11, "*-", color = "purple", label = "BPSK Codificado Simulado - ReedSolomon ($m = 8$, $t = 11$)")

# plt.legend()

plt.title("Taxa de erro de bit BPSK")
plt.xlabel("$E_b/N_0$ (dB)")
plt.ylabel("BER")
plt.grid()
# plt.savefig("BERCurve-geral.pdf", dpi = 00)
```

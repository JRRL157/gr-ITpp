```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


img = cv2.imread("penguin.jpg", cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print(img.shape)

with open("img_data.data", "wb") as f:
    f.write(img.ravel())

with open("img_data.data", "rb") as f:
    teste_leitura = f.read()
teste_leitura = np.frombuffer(teste_leitura, dtype = np.uint8)
```
